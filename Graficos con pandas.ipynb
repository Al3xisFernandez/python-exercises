```python
import pandas as pd
data= pd.read_csv("titanic.csv")
```

```python
data["Age"].plot.hist()
```

```python
data["Age"].plot.hist(bins=20, title="distribucion de edades", legend=True, color="blue")
```

```python
data.groupby("Sex").size().plot.bar()
```

```python
data.groupby("Sex")["Survived"].sum().plot.bar()
```

```python
data.groupby("Sex").size().plot.pie(title="Proporcion de hombres y mujeres")
```

```python
data.groupby("Sex")["Survived"].sum().plot.pie(title="Proporcion de hombres y mujeres entre los supervivientes")
```
